```python
# Code to print board
from IPython.display import clear_output

def display_board(board):
    print('==============='+'     '+'===============')
    print('||   '+ '|   '+ '|   '+'||'+'     '+'||   '+ '|   '+ '|   '+'||')
    print('|| '+ board[0] +' | '+ board[1] +' | '+ board[2] +' ||'+'     '+'|| '+ '1' +' | '+ '2' +' | '+ '3' +' ||')
    print('||   '+ '|   '+ '|   '+'||'+'     '+'||   '+ '|   '+ '|   '+'||')
    print('==============='+'     '+'===============')
    print('||   '+ '|   '+ '|   '+'||'+'     '+'||   '+ '|   '+ '|   '+'||')
    print('|| '+ board[3] +' | '+ board[4] +' | '+ board[5] +' ||'+'     '+'|| '+ '4' +' | '+ '5' +' | '+ '6' +' ||')
    print('||   '+ '|   '+ '|   '+'||'+'     '+'||   '+ '|   '+ '|   '+'||')
    print('==============='+'     '+'===============')
    print('||   '+ '|   '+ '|   '+'||'+'     '+'||   '+ '|   '+ '|   '+'||')
    print('|| '+ board[6] +' | '+ board[7] +' | '+ board[8] +' ||'+'     '+'|| '+ '7' +' | '+ '8' +' | '+ '9' +' ||')
    print('||   '+ '|   '+ '|   '+'||'+'     '+'||   '+ '|   '+ '|   '+'||')
    print('==============='+'     '+'===============')
```

```python
#Code to input names and symbol choices from players
def player_input(playler1_name,playler2_name):
    cond1= True
    while cond1==True:
        sym1 = input(f'Please enter the symbol that {playler1_name} will use in the game - select X or O:')
        if sym1=='X' or sym1=='x' or sym1=='O' or sym1=='o':
            cond1==False
            break
        else:
            print('Invalid selection,please select X or O')
            
    if sym1 == 'X' or sym1 =='x':
        sym2 = 'O'
        sym1 = 'X'
    else:
        sym2 = 'X'
        sym1 = 'O'
    return [sym1,sym2]
```

```python
def place_marker(board, marker, position):
    board[position-1]=marker
```

```python
def space_check(board, position):
    if position<=9 and position>=1 and str(position).isdigit()==True:
        if board[position-1]==' ':
            return True
        else:
            return False
    else:
        return False
```

```python
def win_check(board, mark):
        if board[0]==mark and board[1]==mark and board[2]==mark:
            return True
        elif board[3]==mark and board[4]==mark and board[5]==mark:
            return True
        elif board[6]==mark and board[7]==mark and board[8]==mark:
            return True
        elif board[0]==mark and board[3]==mark and board[6]==mark:
            return True
        elif board[1]==mark and board[4]==mark and board[7]==mark:
            return True
        elif board[2]==mark and board[5]==mark and board[8]==mark:
            return True
        elif board[0]==mark and board[4]==mark and board[8]==mark:
            return True
        elif board[2]==mark and board[4]==mark and board[6]==mark:
            return True
        else:
            return False       
```

```python
#Final game
#Initialize board
print('Welcome Friends to Tic-Tac-Toe')
import time
time.sleep(1)
board=[' ',' ',' ',' ',' ',' ',' ',' ',' ']
display_board_initial(board)
time.sleep(2)
clear_output()
reply = True
while reply == True:  
    answ_ready = input('Ready to play? (Y/N)')
    if answ_ready == 'n' or answ_ready =='N':
        reply = False
        continue
        
    if answ_ready == 'y' or answ_ready =='Y':
        playler1_name = input('Input name of player 1:')
        playler2_name = input('Input name of player 2:')
        sym=player_input(playler1_name,playler2_name)
        print(f'{playler1_name} will use symbol:{sym[0]}')
        print(f'{playler2_name} will use symbol:{sym[1]}')

        print('The postion numbers for the board are as follows')
        initial_board = ['1','2','3','4','5','6','7','8','9']
        display_board_initial(initial_board)
        time.sleep(4)
        clear_output()
        print('The game will start now, Be ready!')
        display_board(board)
   
        win_value=False
        for n in range(0,9):
            while win_value==False:
                if n%2==0:
                    sp_ch = False
                    while sp_ch == False:
                        move1=input(f'{playler1_name} enter position for your move:')
                        while move1.isdigit()==True:
                            sp_ch=space_check(board, int(move1))
                            if sp_ch == True: 
                                clear_output()
                                place_marker(board, sym[0], int(move1))
                                display_board(board)
                                time.sleep(1)
                                break
                            break
                        if sp_ch == False:
                            print('Postion not available for a move')
                    win_value=win_check(board, sym[0])
                    if win_value==True:
                        print(f'{playler1_name} has WON!!!')
                        print(f'{playler1_name} Congratulations :)')
                    break

                
                else:
                    sp_ch = False
                    while sp_ch == False:
                        move2=input(f'{playler2_name} enter position for your move:')
                        sp_ch=space_check(board, int(move2))
                        if sp_ch == True:
                            clear_output()
                            place_marker(board, sym[1], int(move2))
                            display_board(board)
                            time.sleep(1)
                            break
                        if sp_ch == False:
                            print('Postion not available for a move')
                    win_value=win_check(board, sym[1])
                    if win_value==True:
                            print(f'{playler2_name} has WON!!!')
                            print(f'{playler2_name} Congratulations :)') 
                    break
                
        if win_check(board, sym[0])==False and win_check(board, sym[1])==False:
            print('Game Draw!!') 
                
                
        reply = False
    
    

    else:
        print('Invalid input')
        continue
```
